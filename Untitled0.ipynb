```python
import pandas as pd

# Load both files
customers = pd.read_csv("customer_data[1].csv")
payments = pd.read_csv("payment_data[1].csv")

# Preview each one
print("=== CUSTOMERS ===")
print(customers.head())
print("\n=== PAYMENTS ===")
print(payments.head())
```

```python
# Check size of data
print("Customers shape:", customers.shape)
print("Payments shape:", payments.shape)

# Check missing values
print("\n=== MISSING VALUES - CUSTOMERS ===")
print(customers.isnull().sum())

print("\n=== MISSING VALUES - PAYMENTS ===")
print(payments.isnull().sum())

# Check label distribution
print("\n=== CREDIT RISK LABELS ===")
print(customers['label'].value_counts())
```

```python
# Fix missing values in customers
customers['fea_2'].fillna(customers['fea_2'].median(), inplace=True)

# Drop messy columns in payments
payments.drop(columns=['prod_limit', 'report_date'], inplace=True)

# Fix remaining missing values in payments
payments['highest_balance'].fillna(payments['highest_balance'].median(), inplace=True)
payments['update_date'].fillna('unknown', inplace=True)

# Confirm no more missing values
print("Missing in customers:", customers.isnull().sum().sum())
print("Missing in payments:", payments.isnull().sum().sum())
print("Data is clean and ready!")
```

```python
# Merge both tables together on customer ID
df = customers.merge(payments, on='id', how='left')

# Check the result
print("Combined data shape:", df.shape)
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Question 1: How many risky vs good customers?
df_unique = df.drop_duplicates(subset='id')
label_counts = df_unique['label'].value_counts()
print("Good customers (0):", label_counts[0])
print("Risky customers (1):", label_counts[1])

# Visualise it
label_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Good vs Risky Customers')
plt.xlabel('0 = Good | 1 = Risky')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()

# Question 2: Do risky customers have higher overdue payments?
print("\nAverage overdue sum by label:")
print(df.groupby('label')['OVD_sum'].mean())
```

```python
# Chart 2: Overdue amounts - Good vs Risky
df.groupby('label')['OVD_sum'].mean().plot(
    kind='bar', color=['green', 'red'])
plt.title('Average Overdue Amount: Good vs Risky Customers')
plt.xlabel('0 = Good | 1 = Risky')
plt.ylabel('Average Overdue Amount')
plt.xticks(rotation=0)
plt.show()

# Chart 3: Balance comparison
df.groupby('label')['new_balance'].mean().plot(
    kind='bar', color=['green', 'red'])
plt.title('Average Balance: Good vs Risky Customers')
plt.xlabel('0 = Good | 1 = Risky')
plt.ylabel('Average Balance')
plt.xticks(rotation=0)
plt.show()
```
